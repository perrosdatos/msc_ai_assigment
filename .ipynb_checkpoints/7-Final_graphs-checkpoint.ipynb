```python
#==================================
# This code was generated using the prompt and then modified to have desired output:
# "having this sample of my dataframe could you do a plot of the different algorithm 
#   I tested (type) one line for each algorithm in the axis X the size of the graph 
#   in the Y the cost, since there are 10 alternatrives for each size, coul you plot
#   avg cost for each size and type, with some line in vertical to represents std"
#=================================
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt



# Assuming your DataFrame is named df
# Group by 'num_nodes' and 'type' to calculate mean and std
df_grouped = df.groupby(['num_nodes', 'type']).agg(
    avg_cost=('cost', 'mean'),
    std_cost=('cost', 'std')
).reset_index()

# Set up the plot
plt.figure(figsize=(20, 10))

# Iterate over each algorithm type
color = ["blue", "green", "red"]
for idx, algorithm_type in enumerate(df_grouped['type'].unique()):
    data = df_grouped[df_grouped['type'] == algorithm_type]
    plt.errorbar(data['num_nodes'], data['avg_cost'], yerr=data['std_cost'], 
                 label=algorithm_type, capsize=5, marker='o', linestyle='-', color=color[idx])

    
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
# Adding labels and title
plt.xlabel('Number of Nodes', fontsize=18)
plt.ylabel('Average Cost', fontsize=18)
plt.title('Average Cost vs Graph Size for Different Algorithms', fontsize=22)
plt.legend(title='Algorithm Type')
plt.grid(True)

plt.savefig('imgs/avg_costvsgraphsize.png', format='png', dpi=300)

# Display the plot
plt.show()
```

# Fitting an exponential curve

```python
x = list(df_pivot.index.to_list())
y = list(df_pivot["avg_seconds"]["B&B"])
```

```python
import scipy
import numpy as np
params, _ = scipy.optimize.curve_fit(lambda t,a,b: a*np.exp(b*t),  x,  y, p0=(1,0.1))
```

```python
y_fitter = [params[0]*np.exp(params[1]*xx) for xx in x]
f"({params[0]}*e^({params[1]}*x))"
```

```python
# Plotting the actual vs fitted data
plt.figure(figsize=(20,10))
plt.plot(df_pivot.index.to_list(), df_pivot["avg_seconds"]["B&B"], 'o', label="Actual B&B Times", color="blue")
plt.plot(x, y_fitter, '-', label=f"Fitted Exponential Curve ({params[0]}*e^({params[1]}*x))", color="red")
plt.xlabel("Number of Nodes", fontsize=18)
plt.ylabel("Average Execution Time (Seconds)", fontsize=18)
plt.title("Execution Time Growth of B&B Algorithm vs Number of Nodes", fontsize=22)
plt.legend()
plt.grid(True)
plt.savefig('imgs/fittet_graph_showing_time_isgrowing_expnential.png', format='png', dpi=300)
#plt.yscale('log')  # Use log scale to better visualize exponential growth
plt.show()
```
