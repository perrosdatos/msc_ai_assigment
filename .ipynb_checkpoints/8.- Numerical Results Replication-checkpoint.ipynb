# Re-running one simulation

```python
ID_SIMULATION = 70
graph_information = list(filter(lambda item:item["id_simulated"] == ID_SIMULATION, graphs_json))[0]
nodes_list = list(range(graph_information["num_nodes"]))
edges = graph_information["edges"]
```

```python
df[df["id_simulated"] == ID_SIMULATION]
```

```python
#Optimal route
bb_optimal = df[df["id_simulated"] == ID_SIMULATION].query("type == 'B&B'").to_numpy()[0][3]
```

```python
import ga.tsp_ga_din_crossover
# Modify K
ga.tsp_ga_din_crossover.TIMES_EXPLORING = 20
from ga.tsp_ga_din_crossover import TSPGA as GA_P, TIMES_EXPLORING
TIMES_EXPLORING = 10
```

```python
params = {
    "population_prop":1.1, "crossover_proportion":0.8,"mutation_thd":0.3, "exploration_prob":0.2
}
ga = GA_P(nodes_list, edges, **params)
```

```python
tour, cost = ga.run()

print(f"Condition got in the step {ga.steps[-1]} with cost: {ga.costs[-1]} and the tour is: {tour}")
```

# Optimal vs GA-P route
Red edges are edges which are not contanied in the optimal solution

```python
import seaborn as sns
from matplotlib import pyplot as plt
plt.figure(figsize=(20,10))
sns.lineplot(x=ga.steps, y=ga.costs)
plt.title("Step vs Cost", fontsize=28)
plt.xlabel("Step", fontsize=16)
plt.ylabel("Cost", fontsize=16)
plt.show()
```
